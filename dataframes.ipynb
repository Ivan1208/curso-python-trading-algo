```python
import pandas as pd
import numpy as np
import yfinance as yf

df = yf.download (
    'TSLA' ,
    start = '2024-06-01' , 
    end = '2024-06-19' , 
    progress = False
    
)

df # ejemplo de dataframe con datos de yfinance

# podemos exportar este dataframe a archivo csv

df.to_csv('dataframe.csv')

# tambien lo podemos exportar como excel 

df.to_excel('dataframe.xlsx')

# podemos importar archivos csv y excel

import_df = pd.read_csv('dataframe.csv')
import_df

import_dfe = pd.read_excel('dataframe.xlsx')
import_dfe


```
